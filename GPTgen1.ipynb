```python
import numpy as np
import matplotlib.pyplot as plt

# 경계 조건 변화 (예: 매개변수 alpha)
alpha = np.linspace(0, 10, 100)

# 고유값의 실수부 변화 (예: 안정성 지표)
eigenvalues_real = -0.1 * alpha + 1  # 실수부
eigenvalues_imag = 0.5 * np.sin(alpha)  # 허수부

# 안정성 영역 (실수부 < 0)
stability_region = eigenvalues_real < 0

# 그래프 생성
plt.figure(figsize=(10, 6))
plt.plot(alpha, eigenvalues_real, label="Real Part of Eigenvalue", color='blue')
plt.plot(alpha, eigenvalues_imag, label="Imaginary Part of Eigenvalue", color='green', linestyle='--')

# 안정성 영역 음영 처리
plt.fill_between(alpha, eigenvalues_real, 0, where=stability_region, color='red', alpha=0.2, label="Unstable Region")

# 그래프 설정
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.title("Eigenvalue Sensitivity Analysis")
plt.xlabel("Boundary Condition Parameter (α)")
plt.ylabel("Eigenvalue Components")
plt.legend()
plt.grid(True)
plt.show()
```
